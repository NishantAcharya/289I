```python
#!  /usr/bin/env python
import json, sys
handle = open("/Users/klvijeth/Documents/IXP measurement/ixs_202401.jsonl")
decoder = json.JSONDecoder()
objects = []
for line in handle:
    if len(line) > 0 and line[0][0] != "#":
        objects.append(decoder.decode(line))
```

```python
import pandas as pd

df = pd.read_json(json.dumps(objects))

df.head()
```

```python
countries=["TR","RO","SK"]
filtered_df = df[df['country'].isin(countries)]
filtered_df['prefixes']
```
